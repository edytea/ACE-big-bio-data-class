```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example DataFrame
data = pd.DataFrame({"A": [1, 2, 3, 4, 5], "B": [5, 4, 3, 2, 1], "C": [3, 4, 5, 6, 7]})

# Plot using Seaborn
sns.scatterplot(x="A", y="B", data=data)
plt.title("Scatter Plot of A vs B")
plt.xlabel("A")
plt.ylabel("B")
plt.show()

sns.boxplot(data=data)
plt.title("Boxplot of Variables")
plt.show()

sns.histplot(data=data, x="C", bins=5, kde=True)
plt.title("Histogram of C")
plt.xlabel("C")
plt.ylabel("Frequency")
plt.show()
```
