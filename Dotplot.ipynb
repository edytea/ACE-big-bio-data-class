```python
import numpy as np
import matplotlib.pyplot as plt

# Define DNA sequences
seq1 = "ATCGATCGATCGATCG"
seq2 = "ATCGATCGATCGATAG"

# Function to calculate dot matrix
def calculate_dot_matrix(seq1, seq2, window_size):
    dot_matrix = np.zeros((len(seq1), len(seq2)))
    for i in range(len(seq1)):
        for j in range(len(seq2)):
            window1 = seq1[i:i+window_size]
            window2 = seq2[j:j+window_size]
            if len(window1) == len(window2):
                if window1 == window2:
                    dot_matrix[i,j] = 1
    return dot_matrix

# Plot dot matrix
def plot_dot_matrix(dot_matrix, title):
    plt.imshow(dot_matrix, cmap='gray', interpolation='none')
    plt.title(title)
    plt.xlabel('Sequence 2')
    plt.ylabel('Sequence 1')
    plt.show()

# Calculate dot matrix for different window sizes and plot
window_sizes = [1, 2, 3, 4]  # Experiment with different window sizes
for window_size in window_sizes:
    dot_matrix = calculate_dot_matrix(seq1, seq2, window_size)
    plot_dot_matrix(dot_matrix, f'Dot Matrix (Window Size = {window_size})')
```
